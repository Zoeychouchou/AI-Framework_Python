```python
import pandas as pd
train_df = pd.read_csv("house_prices/train.csv",encoding="utf-8")
test_df = pd.read_csv("house_prices/test.csv",encoding="utf-8")
```

```python
#concat axis=0 並在下方, axis=1 並在
total_df = pd.concat([train_df,test_df],axis=0)
total_df
```

```python
#data為處理對象
data = total_df.drop(["SalePrice"],axis=1)
y_train = train_df["SalePrice"]
```

```python
# Solution
med = data.median().drop(["MSSubClass"])
data = data.fillna(med)
```

```python
#One-Hot
data = pd.get_dummies(data)
data = pd.get_dummies(data, columns =["MSSubClass"])
data = data.drop(["Id"], axis=1)
data.head()
```

```python
# 確定無缺失值
s = data.isna().sum()
s[s>0]
```

```python
#先補完 最後才用到，分開x_train,x_test
train_df.shape[0]
#loc: 列標籤， iloc:0,1...
#iloc -> ["第一筆","第二筆"...]
x_train = data.iloc[:train_df.shape[0]]
x_test = data.iloc[train_df.shape[0]:]
x_test
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score
#看score調整params參數
# reg = RandomForestRegressor()
# params = {
#     "n_estimators":range(20,150,10),
#     "max_depth":range(5,20)
# }
# # 換評分標準 更改GridSearchCV scoring
# cv = GridSearchCV(reg, params, scoring="r2", cv=10, n_jobs=4)
# cv.fit(x_train,y_train)
# print(cv.best_params_)
# print(cv.best_score_)
```

```python
reg = RandomForestRegressor(n_estimators=75,max_depth=8)
reg.fit(x_train,y_train)
pre = reg.predict(x_test)
test_id = test_df["Id"]
result = pd.DataFrame({
    "Id":test_id,
    "SalePrice":pre
})
result.to_csv("houseprice_pre.csv", index=False, encoding="utf-8")

```
